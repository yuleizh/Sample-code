```python
## 2&3 group AA counts
def AAcounts_fun(df_VH,df_VL):
    hydroAA1=['A','F','I','L','P','V','W','M','Y']
    hydroAA2=['F','I','L','P','V','W']
    hydropos=['F','I','L','P','V','W','R','K','H']
    philineg=['Q','S','T','N','D','E']
    philiAA1=['Q','S','T','N']
    philiAA2=['Q','S','T','N','Y']
    aromaAA1=['F','W','Y']
    aromaAA2=['F','W','Y','P']
    allAA=['A','C','D','E','F','G','H','I','K','L','M','N','P','Q','R','S','T','V','W','Y']

    AAgroup=[hydroAA1,hydroAA2,hydropos,philineg,philiAA1,philiAA2,aromaAA1,aromaAA2]

    grouptype_VH=[]
    grouptype_VL=[]
    grouptype_Fv=[]
    for k in range(len(AAgroup)):
        typeAA_VH=0
        typeAA_VL=0
        for i in range(len(AAgroup[k])):
            typeAA_VH+=((df_VH==AAgroup[k][i])*1).sum(axis=1)
            typeAA_VL+=((df_VL==AAgroup[k][i])*1).sum(axis=1)
        grouptype_VH.append(list(typeAA_VH))
        grouptype_VL.append(list(typeAA_VL))
        grouptype_Fv.append(list(typeAA_VH+typeAA_VL))

    ## Fv charge
    VH_charge=((df_VH=='R')*1).sum(axis=1)+((df_VH=='K')*1).sum(axis=1)+0.1*((df_VH=='H')*1).sum(axis=1)-((df_VH=='D')*1).sum(axis=1)-((df_VH=='E')*1).sum(axis=1)
    VL_charge=((df_VL=='R')*1).sum(axis=1)+((df_VL=='K')*1).sum(axis=1)+0.1*((df_VL=='H')*1).sum(axis=1)-((df_VL=='D')*1).sum(axis=1)-((df_VL=='E')*1).sum(axis=1)
    Fv_charge=VH_charge+VL_charge
    Fv_asys=VH_charge*VL_charge

    ## Fv hydrophobicity
    hydro_value=[0.499,0.442,0.442,0.379,0.377,0.324,0.237,0.21,0.179,0.115,0,-0.051,-0.142,-0.218,-0.250,-0.265,-0.336,-0.458,-0.473,-0.501]
    hydro_AA=['F','I','L','Y','W','M','V','P','C','A','G','T','S','K','Q','N','H','E','D','R']
    hydrophobicity=pd.DataFrame(hydro_value,hydro_AA)
    hydrophobicity_VH=0
    hydrophobicity_VL=0

    for i in range(len(hydro_value)):
        hydrophobicity_VH+=((df_VH==hydro_AA[i])*hydro_value[i]).sum(axis=1)
        hydrophobicity_VL+=((df_VL==hydro_AA[i])*hydro_value[i]).sum(axis=1)

    hydrophobicity_Fv=hydrophobicity_VH+hydrophobicity_VL 

    Fv_asys_hydro_abs=abs(hydrophobicity_VH-hydrophobicity_VL) 

    Fv_asys_hydro=hydrophobicity_VH-hydrophobicity_VL

    col_VH=['hydroAA1_VH','hydroAA2_VH','hydropos_VH','philineg_VH','philiAA1_VH','philiAA2_VH','aromaAA1_VH','aromaAA2_VH','charge_VH','hydrophobicity_VH']
    col_VL=['hydroAA1_VL','hydroAA2_VL','hydropos_VL','philineg_VL','philiAA1_VL','philiAA2_VL','aromaAA1_VL','aromaAA2_VL','charge_VL','hydrophobicity_VL']
    col_Fv=['hydroAA1_Fv','hydroAA2_Fv','hydropos_Fv','philineg_Fv','philiAA1_Fv','philiAA2_Fv','aromaAA1_Fv','aromaAA2_Fv','charge_Fv','Fv_asys','hydrophobicity_Fv','asys_hydro_abs_Fv','asys_hydro_Fv']


    grouptype_VH.append(VH_charge)
    grouptype_VH.append(hydrophobicity_VH)
    grouptype_VL.append(VL_charge)
    grouptype_VL.append(hydrophobicity_VL)
    grouptype_Fv.append(Fv_charge)
    grouptype_Fv.append(Fv_asys)
    grouptype_Fv.append(hydrophobicity_Fv)
    grouptype_Fv.append(Fv_asys_hydro_abs)
    grouptype_Fv.append(Fv_asys_hydro)

    ## individual AA count
    all_count_VH=[]
    all_count_VL=[]
    for i in range(len(allAA)):
        all_count_VH.append(((df_VH==allAA[i])*1).sum(axis=1))
        all_count_VL.append(((df_VL==allAA[i])*1).sum(axis=1))

    grouptype_VH=pd.DataFrame(grouptype_VH).T
    grouptype_VH.columns=col_VH
    grouptype_VL=pd.DataFrame(grouptype_VL).T
    grouptype_VL.columns=col_VL
    grouptype_Fv=pd.DataFrame(grouptype_Fv).T
    grouptype_Fv.columns=col_Fv

    prop_Fv_AA_VH=pd.DataFrame(all_count_VH).T
    prop_Fv_AA_VL=pd.DataFrame(all_count_VL).T
    prop_Fv_AA_Fv=pd.DataFrame(all_count_VH).T+pd.DataFrame(all_count_VL).T
    prop_Fv_AA_VH.columns=[s + '_num_VH' for s in allAA]
    prop_Fv_AA_VL.columns=[s + '_num_VL' for s in allAA]
    prop_Fv_AA_Fv.columns=[s + '_num_Fv' for s in allAA]

    BI_prop_Fv=pd.concat([grouptype_VH,grouptype_VL,grouptype_Fv,prop_Fv_AA_VH,prop_Fv_AA_VL,prop_Fv_AA_Fv],axis=1,sort=False)

    return BI_prop_Fv


```
